### namespaces
* A namespace is a collection of currently defined symbolic names along with information about the object that each name references.
* You can think of a namespace as a dictionary in which the keys are the object names and the values are the objects themselves.
* Each key-value pair maps a name to its corresponding object.
* These have differing lifetimes. As Python executes a program, it creates namespaces as necessary and deletes them when theyâ€™re no longer needed.
* Namespaces are used to organize and store objects in a program.
* Python has several different namespaces, including global, local, and built-in namespaces.
* The global namespace is the outermost namespace and contains all the objects that are defined globally in a program.
* The local namespace is created when a function is defined and contains all the objects that are defined within the function.
* The built-in namespace contains all the objects that are defined by the Python interpreter.
* Namespaces can be nested, meaning that a namespace can contain other namespaces.
* Namespaces are used to avoid name collisions between objects in different parts of a program.
* Namespaces are also used to implement data hiding, which means that objects can only be accessed from within the namespace in which they are defined.
* Typically, many namespaces will exist at any given time.

* A variable is only available from inside the region it is created. This is called scope
### types of scope of namespaces in python:


* When you access a variable in Python, the interpreter searches for the variable in the local namespace. If the variable is not found in the local namespace, the interpreter searches for it in the enclosing namespace. If the variable is not found in the enclosing namespace, the interpreter searches for it in the global namespace. If the variable is not found in the global namespace, the interpreter raises a NameError exception.

### local scope
* A variable created inside a function belongs to the local scope of that function, and can only be used inside that function.
* It cannot be accessed anywhere outside the function.

```python
# A variable created inside a function is available inside that function:
def myfunc():
  x = 300
  print(x)
myfunc()
```

### enclosed scope
* Enclosed variable is used in the case of nested functions.
* This keyword works similarly to the global, but rather than global, this keyword declares a variable to point to the variable of an outside enclosing function, in case of nested functions.

```python
def myfunc():
    x=300
    print(x)
    def g():
        y=400
        print(y)
```

```python
myfunc()
```

### global scope
* A variable created in the main body of the Python code is a global variable and belongs to the global scope.
* Global variables are available from within any scope, global and local.

```python
# A variable created outside of a function is global and can be used by anyone:

x = 300
def myfunc():
  print(x)
myfunc()
```

### Global Keyword
* If you need to create a global variable, but are stuck in the local scope, you can use the global keyword.
* The global keyword makes the variable global.

```python
# If you use the global keyword, the variable belongs to the global scope:

def myfunc():
  global x
  x = 300
myfunc()
print(x)
```

### built-in scope
* The global namespace includes functions and other variables defined at the main level of a program.
* The local namespace includes local names inside a function.
* The built-in namespace includes built-in functions and built-in exception names.
* The enclosing (or nonlocal) scope is a special scope that only exists for nested functions. If the local scope is an inner or nested function, then the enclosing scope is the scope of the outer or enclosing function.

```python
# Create a built-in namespace
builtin_namespace = {
    "int": int,
    "str": str,
    "list": list,
    "print": print,
    "len": len
}
# Print the built-in namespace
print(builtin_namespace)
```
